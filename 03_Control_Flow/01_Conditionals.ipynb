```python
## If statement
age = 10
if age >= 10:
    print(age)
elif age == 20 :
    print("Age is 20")
else :
    print("Age is 30")
```

```python
# Nested if condition

name = input("Enter your name : ")
if name == "Kishore":
    print(name)
    age = int(input("Enter your age : "))
    if age >= 0:
        print("You are old enough!")
    else :
        print("You are not even born")
else :
    print("You are not Kishore")
```
