### Sets
Sets are collection of unique items. They are unordered and useful for membership tests, eliminating duplicates and performing mathematical operations.
1. Creating a set
2. Basic operations
3. Mathematical operations

```python
# Creating a set
sample_set = {1, 2, 3, 4, 5}
type(sample_set)
print(sample_set)
```

```python
# Basic operations
sample_set_1 = {1, 1, 2, 3, 4, 5}
print(sample_set_1)

# Adding elements to set
sample_set_1.add(10)
print(sample_set_1)

# Removing elements from set
sample_set_1.remove(5)
# sample_set_1.remove(12)  Throws an error as 12 doesn't exist
print(sample_set_1)

# Removing elements from set without throwing an error if it doesn't exist
sample_set_1.discard(12)
print(sample_set_1)
```

```python
# Pop method : Removes random element

number_set = {6, 1, 2, 3, 4, 5, 5}
print(number_set)

char_set = {'c', 'a', 'b', 'd'}
print(char_set)

string_set = {"Dog", "Ball", "Cat", "Apple"}
print(string_set)

number_set_poped_element = number_set.pop()
print(number_set_poped_element)

char_set_poped_element = char_set.pop()
print(char_set_poped_element)

string_set_poped_element = string_set.pop()
print(string_set_poped_element)
```

```python
sample_set = {"Hello", 1, False, False}
print(sample_set)
sample_set.clear()
print(sample_set)
```

```python
# Mathematical operations
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

# Union
set_union = set1.union(set2)
print(set_union)

# Intersection
set_intersection = set1.intersection(set2)
print(set_intersection)
# set1.intersection_update(set2)

# Difference : Elements in Set1 that are not in Set2
set_difference = set1.difference(set2)
# set1.difference_update(set2)
print(set_difference)

set_difference2 = set2.difference(set1)
print(set_difference2)

# Symmetric difference : Elements that are exclusive to both the sets no intersection
set_common = set1.symmetric_difference(set2)
print(set_common)
```

```python
# subset and superset
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
set3 = {1, 2, 3}

print(set1.issubset(set2))
print(set1.issuperset(set3))
print(set3.issubset(set1))
```

```python
# Unique chracters in the text

text="In this tutorial we are discussing about sets"

unique_char = set()

for char in text:
    unique_char.add(char)

print(unique_char)
print(len(unique_char) - 1)
```

```python
# Unique word in the text

text="In this tutorial we are discussing about sets"
words = text.split()
print(words)
print(type(words))

unique_words = set()

for word in words:
    unique_words.add(word)
    
print(unique_words)
print(len(unique_words))
```
