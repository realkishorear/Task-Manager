### Tuple
Tuples are order collection of elements that are immutable. They are similar to lists but they are immutable.
1. Creating tuples
2. Accessing tuple elements
3. Tuple operation
4. Immutable nature of tuple
5. Tuple methods
6. Packing and unpacking tuples
7. Nested tuples

```python
# Creating tuples

# Method 01
dummy_tuple = ()
type(dummy_tuple)

# Method 02
dummy_tuple_2 = tuple()
type(dummy_tuple_2)
```

```python
# Accessing tuple elements

tpl = ("Kishore", 123, False)
print(tpl)
print(tpl[0])

for element in tpl:
    print("Element : ", element)
```

```python
# Tuple Methods

tpl1 = (1, 2, "Kishore", False, 1)

# counting a specific element in tuple
tpl1.count(1)

# finding index of a specific element (First occurence)
tpl1.index(False)

# packing tuples
tpl2 = 1, False, "Kishore"
type(tpl2)
tpl2

# unpacking tuples
roll, isActive, name = tpl2
print(roll, isActive, name)
```

```python
# Nested tuples

nested_tuples = (('Kishore', 'sample', False), ("Sample", "Nothing", True))
for tuples in nested_tuples:
    for element in tuples:
        print(element, end=" ")
    print("")
```
