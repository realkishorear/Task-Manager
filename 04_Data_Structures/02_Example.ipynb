### Real-World Examples Using Lists in Python
Lists are one of the most commonly used data structures in Python, thanks to their versatility and ease of use. Here are several practical examples that illustrate their use in real-world scenarios

Example 1. Manage A To Do List
Create a To Do List To Keep Track OF Tasks

```python
tasks = []
active = True

def display():
    for task in tasks:
        if task:
            print(task)

while active:
    print("1.Add item\n2.Remove Item\n3.Display Item\n4.Exit menu")
    choice = input("Enter your choice : ")
    if choice == "1":
        task = input("Enter the task you want add : ")
        tasks.append(task)
    elif choice == "2":
        display()
        task = input("Enter the task for you want to remove : ")
        for t in tasks:
            if t == task:
                t = None
        print("Task removed sucessfully")
    elif choice == "3":
        display()
    else:
        active = False
        break
```
