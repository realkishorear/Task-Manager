```python
# Introduction to lists
lst = []
print(type(lst))
```

```python
# Creating lists
names = ["Kishore", "Ranjith", "Alwin", 1, 2, 3]
mixed_elements = [False, "Kishore", 3.24, 1, 3, 4]
print(names)
print(mixed_elements)
```

```python
# Accessing elements in Lists
fruits = ["Apple", "Orange", "Grapes", "Banana"]
print(fruits[0])
# fruits[start:end] (sublists or sub-array)
print(fruits[1:3])
```

```python
# Modifying the list elements
names = ["Sam", "Ram", "Dam"]
names[0] = "Ranjith"
print(names)
```

```python
# List Slicing
fruits[1:] = "Kishore"
print(fruits)

fruits[1:] = ["A", "B", "C"]
print(fruits)
```

```python
# Lists methods

# Append Method
sampleList = [1, 2, 3, 4]
sampleList.append(5) #add elements to the last
print(sampleList)
```

```python
# Remove Method
sampleList = [1, 1, 2, 3, 4]
sampleList.remove(1) #removes the first occurence of the element
print(sampleList)
```

```python
# pop method
sampleList = [1, 2, 3, 4]
poped_element = sampleList.pop() #removes and returns the last element
print(poped_element)
```

```python
# index method
numbers = [1, 2, 3, 4, 1, 5]
index = numbers.index(5) #gets the first occuring index of the element
print(index)
```

```python
# Insert element at a particular index not replace
numbers = [1, 2, 3, 4, 5]
numbers.insert(2, "Kishore")
print(numbers)

# Counting elements
print(numbers.count(1))
```

```python
# Sort element in ascending order
numbers = [1, 3, 2, 4, 5]
numbers.sort()
numbers
```

```python
# Reverse elements
numbers = [5, 4, 3, 2, 1]
numbers.reverse()
numbers
```

```python
# Clear
numbers = [1, 2, 3, 4, 5]
print(numbers)
numbers.clear()
numbers
```

```python
# Slicing : numbers(start:stop) start --> included and stop --> not included 
numbers = [1, 2, 3, 4, 5, 6]
print(numbers[0:6])
print(numbers[2:4])
print(numbers[-2:-1])
print(numbers[-6:-1])
print(numbers[::-1]) #reversing elements
print(numbers[::-2]) #skipping 2 elements at a time from last
```

```python
#iterating over a list
numbers = [1, 2, 3, 4, 5]
for number in numbers:
    print(number)
```

```python
# iterating over index and element
numbers = [1, 2, 3, 4, 5, 6]
for index, number in enumerate(numbers):
    number *= abs(3.14/(22/7))
    print(index, number)
```

```python
# List comphrension : 

# Way 01
lst = []
for x in range(5):
    lst.append(x ** 2)
print(lst)

# Way 02
sample = [x ** 2 for x in range(5)]
sample
```

```python
# List comphrension with conditionals
even_numbers = []
for x in range(10):
    if x % 2 == 0:
        even_numbers.append(x)

# list = [return loop conditions]
even_numbers = [x for x in range(10) if x % 2 == 0]
print(even_numbers)
```

```python
# Nested list comphrehension
words = ['Apple', 'Orange', 'Pear', 'Banana']
lenghts = [len(word) for word in words]
lenghts
```
