```python
# Declarations

empty_dict = {}
sample_dict = dict()

print(type(empty_dict))
print(type(sample_dict))
```

```python
sampleData = {
    "name" : "Kishore",
    "age" : 23
}

print(sampleData.get('age', 'Not available bitch!')) #slightly slower but don't throw keyError
print(sampleData['age']) 

sampleData["sex"] = "Male"
print(sampleData)

del sampleData["name"]
print(sampleData)
```

```python
# Dictionary Methods

dict_sample = {
    "name" : "Kishore",
    "initial" : 'A R',
    "sex" : "Male",
    "phone" : None
}

keys = dict_sample.keys()
print(keys)

values = dict_sample.values()
print(values)

pairs = dict_sample.items()
print(pairs)
```

```python
# Deep copy and shallow copy

# Shallow copy
dict_sample = {
    "name" : "Kishore",
    "age" : 23
}

# Shallow copy : Points same object with 2 different reference
sample_dict = dict_sample
sample_dict_1 = dict_sample.copy()

dict_sample["sample"] = "Lol"
print(dict_sample)

sample_dict["lol"] = "Sample"
print(sample_dict)

# Deep copy : Points to totally different objects which is copy of original
import copy

sample_dict_2 = copy.deepcopy(sample_dict)
sample_dict_2["random"] = "LKS"

print(sample_dict_2)
print(sample_dict)
```

```python
# Iterating over dictionary

sample_dik = {
    "name" : "Kishore",
    "age" : 23,
    "gender" : "Male"
}

for keys in sample_dik.keys():
    print(f"Keys are {keys}")

for values in sample_dik.values():
    print(f"Values are {values}")

for keys, values in sample_dik.items():
    print(f"{keys} : {values}")
```

```python
# Dictionary comprehension
square_dict = { x : x**x for x in range(5) }
print(square_dict)
```
