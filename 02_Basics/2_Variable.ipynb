### Variables
- Variables are unit of memory where you can store data.
- The memory is created once the value is assigned.
### Naming convention
- Variable name should be descriptive
- They must start with '_' or letters
- Variable names are case sensitive

```python
string_variable = "sample string"
integer_variable = 10
float_variable = 10.01
boolean_variable = False
```

```python
# Used to print type of the variable
print(type(float_variable))
```

```python
# Type conversion
sample_float_value = 10.02 #float value
print(int(sample_float_value))
```

```python
# input : Pure input is available only for string. You can type cast and use for everything else.

sample_string = input("Enter the string : ")
age = int(input("What is your age ?"))

print(sample_string)
print(age)
```
